HYPERPARTISAN DATASET

```python
hyperpartisan_dataset = load_dataset("hyperpartisan_news_detection", "bypublisher", "/data") #Not stripped of headers and footers
hyperpartisan_dataset.keys()
```

```python
seed = 32
```

```python
hyperpartisan_train = hyperpartisan_dataset['train']
hyperpartisan_valid = hyperpartisan_dataset['validation']

hyperpartisan_train = hyperpartisan_train.shuffle(seed = seed)
hyperpartisan_valid = hyperpartisan_valid.shuffle(seed = seed)
```

```python
eval_test_split = hyperpartisan_valid.train_test_split(test_size = 0.5)

hyperpartisan_eval = eval_test_split['train']
hyperpartisan_test = eval_test_split['test']
```

```python
func_train = AssessData(hyperpartisan_train['text'])
func_eval = AssessData(hyperpartisan_eval['text'])
func_test = AssessData(hyperpartisan_test['text'])
```

```python
func_train._create_distribution()
```

```python
func_eval._create_distribution()
```

```python
func_test._create_distribution()
```

```python
func_train._visualise(), func_eval._visualise(), func_test._visualise()
```

```python
distribution = func_train._distribution

hyperpartisan_distribution = {"seed": seed,
                            "train": func_train._distribution,
                            "eval": func_eval._distribution,
                            "test": func_test._distribution}
```
