```python
parentdir = "Datasets"

train =parentdir+"/20news-bydate"+"/20news-bydate-train"
test = parentdir+"/20news-bydate"+ "/20news-bydate-train"
```

```python
from utils import PrepareCorpus,AssessData
from sklearn.model_selection import train_test_split
from datasets import Dataset,load_dataset
```

```python
news_group_train = PrepareCorpus(train)
news_group_test = PrepareCorpus(test)
```

```python
news_group_train_corpus = news_group_train._corpus()
news_group_test_corpus = news_group_test._corpus()

train_list_of_strings = []; test_list_of_strings = []

for x in news_group_train_corpus.values():
    train_list_of_strings = train_list_of_strings + x 

for x in news_group_test_corpus.values():
    test_list_of_strings = test_list_of_strings + x
```

```python
#Assess train and test data

train_twenty_newsgroup = AssessData(train_list_of_strings)
test_twenty_newsgroup = AssessData(test_list_of_strings)
```

```python
train_twenty_newsgroup._create_distribution()
train_twenty_newsgroup._visualise()

test_twenty_newsgroup._create_distribution()
test_twenty_newsgroup._visualise()
```
